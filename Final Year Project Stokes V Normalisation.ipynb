```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load Doppler & polarity corrected Stokes V profiles

df = np.load("C:/DCU_final_year_project/polarity_correction/ItoQUVcorrected_CBJTML_polarity_corrected.npy")
```

```python
df.shape
```

```python
# Generate Stokes V plot at sample pixel
#y can be in range 0-2554, x can be 0-899
y_test, x_test = 900, 20

sample_plot = df[:, y_test, x_test]
plt.plot(sample_plot)
plt.title(f"Doppler Corrected Profile at (y = {y_test}, x = {x_test}) in original dataset")

plt.show()

# Find and show min and max values of Stokes V plot
max_value = np.nanmax(sample_plot)
min_value = np.nanmin(sample_plot)

print(max_value)
print(min_value)
```

```python
# Set the wavelength axis for plot
wavelengths = np.linspace(0, 181, num = 182)



def normalise(arr):
    # This function assumes the highest magnitude value is positive as
    # the Stokes V plots have already been polarity-corrected
    max_value = np.nanmax(arr)
    
    # Divide array by its max value to set its peak to 1
    arr = [i / max_value for i in arr]
    return arr

test = normalise(sample_plot)
y_test, x_test = 900, 20
#plt.plot(wavelengths, sample_plot, label='Input at [y = ' + str(y_test) + ', x = ' + str(x_test) + ']')
plt.plot(wavelengths, test)
plt.title(f"Normalised Profile at (y = {y_test}, x = {x_test}) in original dataset")
plt.hlines(0, 0, 181, linestyle = '--', color = 'gray')
plt.hlines(0.9, 0, 181, linestyle = '--', color = 'gray')
plt.hlines(-0.9, 0, 181, linestyle = '--', color = 'gray')
plt.show()
```

```python
n_lambda, n_y, n_x = df.shape
normalised_stokes_v = np.empty_like(df, dtype = df.dtype)

for y in range(n_y):
    for x in range(n_x):
        normalised_stokes_v[:, y, x] = normalise(df[:, y, x])
    print(f"Processed y = {y + 1}/{n_y}")
```

```python
# Generate Stokes V plot at sample pixel
#y can be in range 0-2554, x can be 0-899
y_test, x_test = 1220, 130
sample_plot = normalised_stokes_v[:, y_test, x_test]
plt.plot(sample_plot)
plt.hlines(0, 0, 181, linestyle = '--', color = 'gray')
plt.hlines(0, 0, 181, linestyle = '--', color = 'gray')
plt.hlines(1, 0, 181, linestyle = '--', color = 'gray')
plt.hlines(-1, 0, 181, linestyle = '--', color = 'gray')
plt.title(f"Normalised Profile at (y = {y_test}, x = {x_test})")
plt.show()

# Find and show min and max values of Stokes V plot
max_value = np.nanmax(sample_plot)
min_value = np.nanmin(sample_plot)

print(f"Maximum point = {max_value}")
print(f"Minimum point = {min_value}")
```

```python
# Save the normalised profiles
np.save('C:/DCU_final_year_project/normalised_profiles/ItoQUVcorrected_CBJTML_normalised_stokes_v', normalised_stokes_v)
```
